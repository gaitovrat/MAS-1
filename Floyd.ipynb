```python
import csv
import sys
```

```python
N = 34
FILENAME = 'KarateClub.csv'
```

```python
def print_data(d: list[list[int]]) -> None:
    for i in range(len(d)):
        for j in range(len(d[i])):
            print(d[i][j], end=" ")
        print('')
```

```python
data = [[0] * N for _ in range(N)]
with open(FILENAME, "r") as fd:
    reader = csv.reader(fd, delimiter=';')
    for line in reader:
        i = int(line[0]) - 1
        j = int(line[1]) - 1
        data[i][j] = data[j][i] = 1
        
print_data(data)
```

```python
def floyd(m: list[list[int]]) -> list[list[int]]:
    n = len(m)
    dist = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            dist[i][j] = m[i][j] if m[i][j] != 0 else sys.maxsize
    
    for i in range(n):
        dist[i][i] = 0

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][j] > (dist[i][k] + dist[k][j]):
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist
```

```python
dist = floyd(data)
print_data(dist)
```

```python
prumerna_vzdalenost = 0
n = len(dist)

for i in range(n):
    for j in range(i + 1, n):
        prumerna_vzdalenost += dist[i][j]

prumerna_vzdalenost *= (2 / (n * (n - 1)))
prumerna_vzdalenost
```

```python
prumer = 0
n = len(dist)

for i in range(n):
    for j in range(n):
        prumer = max(prumer, dist[i][j])

prumer
```

```python
n = len(dist)
closeness = [0 for _ in range(n)]

for i in range(n):
    sum = 0
    for j in range(n):
        sum += dist[i][j]
    closeness[i] = n / sum

for i in range(len(closeness)):
    print(f'{i}: {closeness[i]:.7f}')
```
