```python
import numpy as np
import pandas as pd

from mas import adjacency_matrix, floyd
```

```python
FILENAME = "../data/KarateClub.csv"
```

```python
data = adjacency_matrix(FILENAME)
pd.DataFrame(data)
```

```python
dist = floyd(data)
pd.DataFrame(dist)
```

```python
prumerna_vzdalenost = 0
n = len(dist)

for i in range(n):
    for j in range(i + 1, n):
        prumerna_vzdalenost += dist[i][j]

prumerna_vzdalenost *= 2 / (n * (n - 1))
float(prumerna_vzdalenost)
```

```python
prumer = 0
n = len(dist)

for i in range(n):
    for j in range(n):
        prumer = max(prumer, dist[i][j])

int(prumer)
```

```python
n = dist.shape[0]
closeness = np.zeros((n,), dtype=np.float64)

for i in range(n):
    sum = 0
    for j in range(n):
        sum += dist[i][j]
    closeness[i] = n / sum

pd.DataFrame(closeness)
```
