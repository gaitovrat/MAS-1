```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from mas import adjacency_list, adjacency_matrix
```

```python
FILENAME = "../data/KarateClub.csv"
```

# Adjacency Matrix

```python
matrix = adjacency_matrix(FILENAME)
matrix
pd.DataFrame(matrix)
```

```python
N = matrix.shape[0]
stupen_vrcholu = np.zeros((N,), dtype=int)
for i in range(N):
    stupen_vrcholu[i] = sum(matrix[i])

min_stupen_vrcholu = min(stupen_vrcholu)
max_stupen_vrcholu = max(stupen_vrcholu)
prumer_stupen_vrcholu = sum(stupen_vrcholu) / N

print(f"min: {min_stupen_vrcholu}")
print(f"max: {max_stupen_vrcholu}")
print(f"prumer: {prumer_stupen_vrcholu:.2f}")
```

```python
plt.hist(stupen_vrcholu, edgecolor="black", bins=max_stupen_vrcholu)
plt.show()
```

```python
sum_stupen_vrcholu = sum(stupen_vrcholu)
relativni = [stupen / sum_stupen_vrcholu for stupen in stupen_vrcholu]

plt.hist(relativni, edgecolor="black", bins=max_stupen_vrcholu)
plt.show()
```

# Adjacency List

```python
adj_list = adjacency_list(FILENAME)

pd.DataFrame(adj_list)
```

```python
N = max([max(*el) for el in adj_list])
stupen_vrcholu = np.zeros((N,), dtype=int)
for i in adj_list:
    stupen_vrcholu[i[0] - 1] += 1
    stupen_vrcholu[i[1] - 1] += 1

min_stupen_vrcholu = min(stupen_vrcholu)
max_stupen_vrcholu = max(stupen_vrcholu)
prumer_stupen_vrcholu = sum(stupen_vrcholu) / N

print(f"min: {min_stupen_vrcholu}")
print(f"max: {max_stupen_vrcholu}")
print(f"prumer: {prumer_stupen_vrcholu:.2f}")
```

```python
plt.hist(stupen_vrcholu, edgecolor="black", bins=max_stupen_vrcholu)
plt.show()
```

```python
sum_stupen_vrcholu = sum(stupen_vrcholu)
relativni = [stupen / sum_stupen_vrcholu for stupen in stupen_vrcholu]

plt.hist(relativni, edgecolor="black", bins=max_stupen_vrcholu)
plt.show()
```
